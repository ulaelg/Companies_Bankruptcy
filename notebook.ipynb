#### 4. Data Preparation:

#### Imbalanced data :

##### Split:

+ Split data to X_train, y_train, X_test and y_test :

```python
target ="bankrupt"
y=df[target]
X=df.drop(columns=target)
X_train, X_test, y_train, y_test= train_test_split(X,y, random_state=42, test_size=.2)
print("X_train shape : ",X_train.shape)
print("X_test shape : ",X_test.shape)
print("y_train shape : ",y_train.shape)
print("y_test shape : ",y_test.shape)
```

##### Resamble :

>After we split the data into `training` and `testing` sets, we need to address the `class imbalance` observed during our `EDA`, as it can negatively impact the performance of our future models. There are various methods to address an imbalanced dataset, such as `oversampling` and `undersampling`. In the next step, we will create multiple versions of our dataset using these techniques and build models to evaluate their performance. We will then choose the best version based on the model's performance.

+ Oversampling :

```python
over_sampler = RandomOverSampler(random_state=42)
X_train_over, y_train_over = over_sampler.fit_resample(X_train,y_train)
print(X_train_over.shape)
X_train_over.head()
```

+ Undersampling :

```python
under_sampler = RandomUnderSampler(random_state=42)
X_train_under, y_train_under = under_sampler.fit_resample(X_train,y_train)
print(X_train_under.shape)
X_train_under.head()
```

+ Build Model:

>Create three models: `model_reg`, `model_under` and `model_over`:

```python
# regular model  fit on X_train and y_train
model_reg=make_pipeline(SimpleImputer(strategy="median"),
                        DecisionTreeClassifier(random_state=42))
model_reg.fit(X_train,y_train)

# undersample model  fit on X_train_under and y_train_under
model_under=make_pipeline(SimpleImputer(strategy="median"),
                        DecisionTreeClassifier(random_state=42))
model_under.fit(X_train_under,y_train_under)

# oversampling model  fit on X_train_over and y_train_over
model_over=make_pipeline(SimpleImputer(strategy="median"),
                        DecisionTreeClassifier(random_state=42))
model_over.fit(X_train_over,y_train_over)
                
```

+ Evaluate :

```python
#original data 
print("Original dataset :")
print(f"Training accurary is {round(model_reg.score(X_train,y_train),2)}")
print(f"Test accurary is {round(model_reg.score(X_test,y_test),2)}")

#over_sampler
print("Over sampler :")
print(f"Training accurary is {round(model_over.score(X_train,y_train),2)}")
print(f"Test accurary is {round(model_over.score(X_test,y_test),2)}")

#under sampler 
print("Under sampler :")
print(f"Training accurary is {round(model_under.score(X_train,y_train),2)}")
print(f"Test accurary is {round(model_under.score(X_test,y_test),2)}")
```

Plot a confusion matrix that shows how the best model performs.

```python
# Predictions
y_pred_under = model_under.predict(X_test)
y_pred_over = model_over.predict(X_test)

# Confusion Matrices
cm_under = confusion_matrix(y_test, y_pred_under)
cm_over = confusion_matrix(y_test, y_pred_over)

# Plotting
fig, ax = plt.subplots(1, 2, figsize=(12, 5))

ConfusionMatrixDisplay(cm_under).plot(ax=ax[0], colorbar=False)
ax[0].set_title("Confusion Matrix for Undersampled Model")

ConfusionMatrixDisplay(cm_over).plot(ax=ax[1], colorbar=False)
ax[1].set_title("Confusion Matrix for Oversampled Model")


plt.show()
```

**SUMMARY :**

> + **Undersampled Model:** This model is more effective in identifying bankrupt companies (higher TP and lower FN). However, it also produces more false positives, meaning it incorrectly labels more non-bankrupt companies as bankrupt.
> + **Oversampled Model:** This model is more effective in identifying non-bankrupt companies (higher TN and lower FP). It makes fewer mistakes in predicting non-bankrupt companies but is less effective in identifying bankrupt companies compared to the undersampled model.

+ If the primary goal is to minimize the risk of missing bankrupt companies (i.e., lower false negatives are more critical), the undersampled model may be preferred despite its higher false positive rate.
+ If the primary goal is to minimize false alarms and ensure that non-bankrupt companies are correctly identified, the oversampled model may be preferred due to its lower false positive rate.

#### 5. Build Model:

##### Baseline:

```python
acc_baseline=y_train_over.value_counts(normalize=True).max()
print("Baseline Accuracy",round(acc_baseline,2))
```

##### Iterate:

```python
#create a pipeline named classifier that contains a SimpleImputer transformer and a RandomForestClassifier Predictor 
classifier=make_pipeline(SimpleImputer(),
                         RandomForestClassifier(random_state=42))
                        
#perform cross validation with classifier 
cv_acc_scores=  cross_val_score(classifier,X_train_over,y_train_over,cv=5,n_jobs=-1)
cv_acc_scores
```

> In order to get the best performance from the model, we need the tune its Hyperparameter:

+ Hyperparameters tuning:

```python
# Create a dictionary with the range of hyperparameters:
params = {
    "simpleimputer__strategy": ['mean', 'median'],  
    "randomforestclassifier__n_estimators": range(25, 100, 25),  
    "randomforestclassifier__max_depth": range(10, 50, 10)
}

# Create a GridSearchCV object named model that includes the classifier and the hyperparameters:
model = GridSearchCV(classifier,
                     param_grid=params,
                     cv=5,
                     n_jobs=-1,
                     verbose=1)

# Train the model
model.fit(X_train_over, y_train_over)
```

+ Extract and visualize the cross-validation results from the model:

```python
# Extract the results and lead them into a Dataframe :
cv_results=pd.DataFrame(model.cv_results_)
#display first five rows 
cv_results.head()
```

```python
#create a function
def Time_vs_estimator(ax,depth):
    mask=cv_results['param_randomforestclassifier__max_depth']==depth
    #plot fit time vs n_estimators
    ax.plot(cv_results[mask]["param_randomforestclassifier__n_estimators"],
             cv_results[mask]['mean_fit_time'])
    ax.set_title(f'Training Time vs. Estimators (max_depth={depth})')
    ax.set_xlabel('Number of Estimators')
    ax.set_ylabel('Mean Fit Time [seconds]')
fig,ax=plt.subplots(2,2,figsize=(14, 10))
for ax, depth in [(ax[0,0],10),(ax[1,0],20),(ax[0,1],30),(ax[1,1],40)]:
    Time_vs_estimator(ax,depth)
plt.show()
```

> In the figure above, we observe that as the depth of the random forest increases, the time also increases, we also notice that the imputation strategy has an impact on the training time.

```python
#Extract the best hyperparameter froom model 
model.best_params_
```

```python
#Extract the best score from model 
model.best_score_
```

+ Evaluate:

```python
#calculate the training and test accuracy scores:
acc_train=model.score(X_train_over,y_train_over)
acc_test=model.score(X_test,y_test)
#display result
print("Training accuracy :",round(acc_train,2))
print("Test accuracy : ",round(acc_test,2))
```

```python
#plot confusion matrix
ConfusionMatrixDisplay.from_estimator(model,X_test,y_test)
```
